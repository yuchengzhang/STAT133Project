**In this cell we've downloaded the data into the ../data/raw directory, and renamed it bio_data.soft.gz**

```python
%%R

data_url = 'ftp://ftp.ncbi.nlm.nih.gov/geo/datasets/GDS4nnn/GDS4471/soft/GDS4471.soft.gz'
bio_dir = '../data/raw/bio_data.soft.gz'

download.file(data_url, destfile=bio_dir)
```

**Load Biobase and GEOquery packages of bioconductor**

* Biobase comes with bioconductor
* we need Biobase to use GEOquery which opens up our file for us

```python
%%R

library(Biobase)
library(GEOquery)
```

**Open file with getGEO method of GEOquery**

```python
%%R

data = getGEO(filename=bio_dir)
```

**Make sure the data looks right**

```python
%%R

print(Table(data))
```
